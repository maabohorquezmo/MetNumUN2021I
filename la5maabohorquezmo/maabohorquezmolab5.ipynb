### Periodo de Tiempo: 6 Meses

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        
        # six months
        self.SetStartDate(2020, 9, 1)
        self.SetEndDate(2021, 2, 1)
        
        #set starting cash $1 000 000
        self.SetCash(1000000)

        # Select stocks and add in DataNormalizationMode = Raw
        self.amzn = self.AddEquity("AMZN", Resolution.Minute)
        self.amzn.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.pfe = self.AddEquity("PFE", Resolution.Minute)
        self.pfe.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.gm = self.AddEquity("GM", Resolution.Minute)
        self.gm.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.msft = self.AddEquity("MSFT", Resolution.Minute)
        self.msft.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.dpw.de = self.AddEquity("DPW.DE", Resolution.Minute)
        self.dpw.de.SetDataNormalizationMode(DataNormalizationMode.Raw)
        

    def OnData(self, data):
    # regular captal distribution
        if not self.Portfolio.Invested:
            self.SetHoldings("AMZN", 0.2)
            self.SetHoldings("PFE", 0.2)
            self.SetHoldings("GM", 0.2)
            self.SetHoldings("MSFT", 0.2)
            self.SetHoldings("GOOG", 0.2)
            
            
            # print Data
            self.Log("Quantity:" + str(self.Portfolio['AMZN'].Quantity));
            self.Log("Quantity:" + str(self.Portfolio['PFE'].Quantity));
            self.Log("Quantity:" + str(self.Portfolio['GM'].Quantity));
            self.Log("Quantity:" + str(self.Portfolio['MSFT'].Quantity));
            self.Log("Quantity:" + str(self.Portfolio['DPW.DE'].Quantity));
```
